```python
import csv

# Create a class AdAnalytics
class AdAnalytic:
    def __init__(self, date, tag, sub_type, ad_type, rating, cost, predicted_ctr):
        self.date = date
        self.tag = tag
        self.sub_type = sub_type
        self.type = ad_type
        self.rating = rating
        self.cost = cost
        self.predicted_ctr = predicted_ctr
        # Unique ID for each instance
        self.unique_id = id(self)
    
    # __Str__ method for string representation
    def __str__(self):
        return f"{self.unique_id},{self.date},{self.tag},{self.sub_type},{self.type},{self.rating},{self.cost},{self.predicted_ctr}"

# Subclass for NativeAdAnalytic   
class NativeAdAnalytic(AdAnalytic):
    def __init__(self, date, tag, sub_type, rating, cost, predicted_ctr):
        # Intialize with ad_type specified
        super().__init__(date, tag, sub_type, "Native", rating, cost, predicted_ctr)
        
# Subclass NonNativeAdAnalytic
class NonNativeAdAnalytic(AdAnalytic):
    def __init__(self, date, tag, sub_type, rating, cost, predicted_ctr):
        #  Intialize with ad_type specified
        super().__init__(date, tag, sub_type, "non-Native", rating, cost, predicted_ctr)
        
# Creating three additional subclasses
class InApp(NativeAdAnalytic):
    def __init__(self, date, tag, rating, cost, predicted_ctr):
        #Initialize using specified sub_type
        super().__init__(date, tag, "InApp", rating, cost, predicted_ctr)
        
        
class Banner(NonNativeAdAnalytic):
    def __init__(self, date, tag, rating, cost, predicted_ctr):
        super().__init__(date, tag, "Banner", rating, cost, predicted_ctr)
        
        
class Interstitial(NonNativeAdAnalytic):
    def __init__(self, date, tag, rating, cost, predicted_ctr):
        super().__init__(date, tag, "Interstitial", rating, cost, predicted_ctr)

# Testing the code
BannerAdAnalytic = Banner("2022-11-15", "tagA", 2, 641, 86.06)
print(str(BannerAdAnalytic))


```

```python
# Line Plot
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 
# Assuming 'your_data.csv' is the name of your CSV file
file_path = 'data.csv'
 
# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path)
 
# Convert 'date' to datetime if not already in datetime format
df['date'] = pd.to_datetime(df['date'])
 
# Sort the DataFrame by 'date'
df = df.sort_values(by='date')
 
# Extract the first 1000 values
df_subset = df.head(1000)
 
# Plotting the line graph for 'date' and 'predicted_ctr' with seaborn
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
 
# Using seaborn lineplot for 'predicted_ctr'
sns.lineplot(x='date', y='predicted_ctr', data=df_subset, label='Predicted CTR', marker='o', linestyle='-', color='blue')
 
# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Line Graph for Predicted CTR to date')
plt.legend()
 
# Set the y-axis range (adjust the values as needed)
plt.ylim(0,20)
 
# Get a Grid for the graph
plt.grid(True)
 
# Show the plot
plt.show()
```

```python
# Pie chart

import pandas as pd
import matplotlib.pyplot as plt
 
# Load your dataset 
Analytics_data = pd.read_csv('data.csv')  
 
# Assuming your dataset has a column named 'Rating', you can proceed with the pie chart
rating_counts = Analytics_data['rating'].value_counts()
 
# Define custom colors
custom_colors = plt.cm.Set3.colors[:len(rating_counts)]
 
# Plotting the pie chart with custom colors
plt.figure(figsize=(8, 8))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
plt.title('Distribution of Ratings')
plt.show()
```

```python
#Heat Maps
import seaborn as sns
import matplotlib.pyplot as plt
 
# Assuming your dataset is named Analytics_data
plt.figure(figsize=(12, 8))
 
# Pivot table to reshape the data for the heatmap
heatmap_data = Analytics_data.pivot_table(index='type', columns='rating', values='cost', aggfunc='mean')
 
# Plot the heatmap
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', linewidths=.5, fmt=".1f")
 
# Add title and labels
plt.title('Heatmap of Cost by type and rating')
plt.xlabel('Rating')
plt.ylabel('type')
 
# Show the plot
plt.show()
```
